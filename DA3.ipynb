````python
page_posts = pd.read_table('page_85409903065_2019_02_16_22_16_30_fullstats.tab')
comments = pd.read_table('page_85409903065_2019_02_16_22_16_30_comments.tab')
````

````python
page_posts
````

````python
comments
````

````python
def word_present(text, query):
    import re
    text = str(text).lower()
    query = str(query).lower()
    
    tokens = re.findall(r"[\w']+|[.,!?;$@#]", text)
    if query in tokens:
        return 1
    return 0
````

````python
page_posts['thank'] = page_posts['post_message'].apply(word_present, args=('thank',))
````

````python
comments['thank'] = comments['comment_message'].apply(word_present, args=('thank',))
````

````python
page_posts['thank'].value_counts()
````

````python
page_posts[(page_posts['thank'] == 1)].describe()
````

````python
comments['thank'].value_counts()
````

````python
comments[(comments['thank'] == 1)].describe()
````

**The results indicated that ‘thank’ appeared 13 times in the page posts, and 57 times in the comments.**

```python
import re
from collections import Counter
```

```python
texts = twitter['text'].values.tolist()
```

```python
total_words = Counter()
for text in texts:
    text = text.lower()
    text = text.split(' ')
    newtext = []
    for item in text:
        if 'http' not in item:
            newtext.append(item)
    
    newtext = ' '.join(newtext)
        
    tokens = re.findall(r"[\w']+|[.,!?;$@]", newtext)
    for token in tokens:
        total_words[token] += 1
```

```python
total_words.most_common(100)
```

```python
def wordlist_any_present(text, query):
    import re
    text = str(text).lower()
    newquery = []
    for word in query:
        newquery.append(str(word).lower())
    tokens = re.findall(r"[\w']+|[.,!?;$@#]", text)
    
    for word in newquery:
        if word in tokens:
            return 1
    return 0
```

I did a sanity check by looking at the most frequent keywords, and found that several keywords are actually names of celebrities.
And "ar condicionado" actually means air conditioner in Spanish,therefore I created two categories based on the following criteria.

```python
twitter['celebrity'] = twitter['text'].apply(wordlist_any_present, args=(['bts','boechat','fogo'],))
twitter['airconditioner'] = twitter['text'].apply(wordlist_any_present, args=(['condicionado'],))
twitter['AR'] = twitter['text'].apply(wordlist_any_present, args=(['reality','augemented'],))
twitter['IKEA'] = twitter['text'].apply(wordlist_any_present, args=(['ikea'],))
```

```python
def categorise_df(row):
    row['topic'] = 'other'
    if (row['celebrity'] == 1):
        row['topic'] = 'celebrity'
    if (row['airconditioner'] == 1):
        row['topic'] = 'airconditioner'
    if (row['AR'] == 1):
        row['topic'] = 'AR'
    if (row['IKEA'] == 1):
        row['topic'] = 'IKEA'
    return row
```

```python
tweets = twitter.apply(categorise_df, axis=1)
```

```python
sns.countplot(y="topic", data=tweets, palette="Blues");
```

The results indicated that among the four topics ("air conditioner", "celeberity", "AR" and "IKEA"), people talked more about air conditioner and celebrity. Actually not many people are really talking about AR, and nobody mentioned IKEA. It seems that IKEA Place (the AR app of IKEA) has aroused little awareness on twitter.

```python
SA = pd.read_pickle('Jingjing Wang - JingjingWang_EN_completed.pkl')
```

```python
tweets1 = tweets.merge(SA, on=['id','text'])
```

```python
tweets1
```

```python
def categorise_df(row):
    row['sentiment'] = 'neutral'
    if row['neutral'] > 0:
        row['sentiment'] = 'positive'
    if row['neutral'] < 0:
        row['sentiment'] = 'negative'
    return row
```

```python
tweets1['neutral'] = pd.to_numeric(tweets1['neutral'], errors='coerce')
```

```python
tweets2 = tweets1.apply(categorise_df, axis=1)
```

```python
sns.countplot(x="topic", hue ="sentiment", data=tweets2, palette="Blues");
```

The graph indicated that among the posts about AR, most are neutral posts, and there are more positive posts than negative ones. For the posts that are not categorized into a certain topic, there are around same amount of neutral and positive posts, however, there are far less negative posts.
